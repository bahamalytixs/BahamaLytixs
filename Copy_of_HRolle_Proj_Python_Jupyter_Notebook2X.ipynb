```python
def graph_make(dfStck, dfRev, name):
# XXXX
  
    import datetime as dt
    import matplotlib.pyplot as plt
    from matplotlib import style
    import pandas as pd
    import pandas_datareader.data as web
    style.use('ggplot')

    #print(dfStck.head())
    print(dfRev.head())

    ax1 = plt.subplot2grid((8,1), (0,0), rowspan=4, colspan=1)
    plt.xlabel('Date')
    plt.ylabel('Stock_Values')
    plt.title(name)

    ax2 = plt.subplot2grid((8,1), (5,0), rowspan=4, colspan=1, sharex=ax1)
    plt.xlabel('Date')
    plt.ylabel('Rev_Data')
    plt.title(name)

    ax1.plot(dfStck.index, dfStck[1])
    ax2.plot(dfRev.index, dfRev[1])
    #ax2.bar(df.index, df['Volume'])

    plt.show()
```
